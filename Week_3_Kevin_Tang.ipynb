Instance/Object Attributes

The next simplest thing is to "describe" the object or instance. To add descriptions of the object, we need to give it attributes. Lets say we want to have "Departure", "Arrival", "Miles" "Time"

To assign the attributes, we need the syntax def __init__(self, ...). The first argument is always self. We do not need to write self when we create the actual object but we need it in the definition. 

To add atrrributes, we create a object like before, but now we add the attributes like arrival and departure. 

To call, we use the object/instance_name.attribute

```python
class Train_Routes:
  def __init__(self,Departure,Arrival,Miles,Time):
    self.Departure = Departure
    self.Arrival = Arrival
    self.Miles = Miles
    self.Time = Time

NYDC = Train_Routes("New York", "DC", 100, 3)
NYDC.Miles 
```

Object/Instance Methods

Now, we know how to describe an object/instance. But we can also manipulate the data inside. To do so, we can use a instance/object method!

The syntax is similar to that of a attribute but we only need the "self" and arguments for the method. A Method is written like a function!

Here we are creating a method to calculate the miles per hour of a route!

To call the method, we need to use the object/instance_name.method()




```python
class Train_Routes:
  def __init__(self,Departure,Arrival,Miles,Time):
    self.Departure = Departure
    self.Arrival = Arrival
    self.Miles = Miles
    self.Time = Time

  def Miles_per_Hour(self):
    Miles_per_Hour = self.Miles/self.Time 
    print(Miles_per_Hour)

NYDC = Train_Routes("New York", "DC", 100, 3)
NYDC.Miles_per_Hour()
```

Class Attributes:

Sometimes it is useful for us to find ways to describe the class as a whole. Lets say for example that I run Train and want to know not just information about any single route but about all of my routes. In that case, you might want to use a class attribute whereby you find a way to describe the class. 

Lets say for example that I want to know how many miles all of my routes travel! To do that, I would create a class attribute called total_miles. It does not need any special syntax. It gets called like any normal variable. thats how you know that its a class attribute.

You can use it anywhere in the class but after its declared, you need to use Class_Name.Attribute.

In this case, I am updating total_miles each time a new route is added so I know the total amount of miles my routes are. 

To access the attribute, just write Class_Name.Attribute

```python
class Train_Routes:
  total_miles = 0

  def __init__(self,Departure,Arrival,Miles,Time):
    self.Departure = Departure
    self.Arrival = Arrival
    self.Miles = Miles
    self.Time = Time
    Train_Routes.total_miles += self.Miles

  def Miles_per_Hour(self):
    Miles_per_Hour = self.Miles/self.Time 
    print(Miles_per_Hour)

NYDC = Train_Routes("New York", "DC", 100, 3)
Seattle_Portland = Train_Routes("Seattle", "Portland", 500, 8)
print(Train_Routes.total_miles)
```

Class Methods

The last thing to cover in class today is the Class Methods structure. We talked before about how we sometimes want to manipulate the object attribute with object methods. Thus, we sometimes also want to describe Class ATrributes with Class Methods.

Class Methods must be written with the syntax: @classmethod.

It must also use cls as the first argument.

The Class Method odometer simply returns the Class Attribute.

To declare the method, simply write Class_Name.Method()



```python
class Train_Routes:
  total_miles = 0

  @classmethod 
  def odometer(cls):
    return f'{Train_Routes.total_miles} traveled' 

  def __init__(self,Departure,Arrival,Miles,Time):
    self.Departure = Departure
    self.Arrival = Arrival
    self.Miles = Miles
    self.Time = Time
    Train_Routes.total_miles += self.Miles

  def milesperhour(self):
    milesperhour = self.Miles/self.Time 
    print(milesperhour)

NYDC = Train_Routes("New York", "DC", 100, 3)
Seattle_Portland = Train_Routes("Seattle", "Portland", 500, 8)
Train_Routes.odometer()

```

Class Methods and attributes can actually be accessed by instances/methods but instance methods cannot be accessed by classes.

```python
print(Train_Routes.odometer())
print(Seattle_Portland.odometer())
#Train_Routes.milesperhour()
Seattle_Portland.milesperhour()
```

Now we have covered the absolute basics needed to create and work with classes and objects. We can also do some more advanced things with these structures. For example, we can make class and ojbect/instance methods take arguments that are not in the attributes. 

In this case, we made a object attribute called delay that takes an argument, delay and adds it to the time of a route.

```python
class Train_Routes:
  total_miles = 0

  @classmethod 
  def odometer(cls):
    return f'{Train_Routes.total_miles} traveled' 

  def __init__(self,Departure,Arrival,Miles,Time):
    self.Departure = Departure
    self.Arrival = Arrival
    self.Miles = Miles
    self.Time = Time
    Train_Routes.total_miles += self.Miles

  def milesperhour(self):
    milesperhour = self.Miles/self.Time 
    print(milesperhour)
  
  def delay(self, delay):
    self.Time += delay

NYDC = Train_Routes("New York", "DC", 100, 3)
Seattle_Portland = Train_Routes("Seattle", "Portland", 500, 8)
NYDC.delay(1)
NYDC.Time
```

Class Methods can also take arguments. 



```python
class Train_Routes:
  total_miles = 0

  @classmethod 
  def odometer(cls):
    return f'{Train_Routes.total_miles} traveled' 

  @classmethod 
  def yearly_miles(cls, trips_per_year):
    return Train_Routes.total_miles * trips_per_year

  def __init__(self,Departure,Arrival,Miles,Time):
    self.Departure = Departure
    self.Arrival = Arrival
    self.Miles = Miles
    self.Time = Time
    Train_Routes.total_miles += self.Miles

  def milesperhour(self):
    milesperhour = self.Miles/self.Time 
    print(milesperhour)
  
  def delay(self, delay):
    self.Time += delay

NYDC = Train_Routes("New York", "DC", 100, 3)
Seattle_Portland = Train_Routes("Seattle", "Portland", 500, 8)
NYDC.delay(1)
NYDC.Time
Train_Routes.yearly_miles(200)
```

Sometimes we have categories within categories. For example, in the animal kingdom we have multiple levels of organization. Each one could potentially be its own class. But under the biggest category Domain.

In Python, we would represent that through Inheritance. One Class can inherit all of the methods and attributes of another class. 

image.png


For example, lets say I was Train executive and wanted to look at the the sources of profit. I might have a big Train class with different activites below it also as classes.

We inherit by using class_name(class_to_inherit) to receive all of the properties of the previous class. 
We can give the overaching class, Train_Inc with attributes and properties. We can have Train_Routes, and Train_Merchandise inherit all the properties and attributes of Train_Inc. So we can call its attributes and methods. 

```python
class Train_Inc:
  def __init__(self, income, expenditure ):
    self.income = income
    self.expenditure = expenditure
  def profit(self):
    return self.income - self.expenditure

class Train_Routes(Train_Inc):
   def __init__(self, income, expenditure,Departure,Arrival,Miles,Time):
    Train_Inc.__init__(self, income,expenditure)
    self.Departure = Departure
    self.Arrival = Arrival
    self.Miles = Miles
    self.Time = Time
    Train_Routes.total_miles += self.Miles

class Train_Merchandise(Train_Inc):
   def __init__(self, income, expenditure, pieces_sold):
    Train_Inc.__init__(self, income,expenditure)
    self.pieces_sold = pieces_sold 

Tshirts = Train_Merchandise( 10000, 8000, 500)
print(Tshirts.profit())
print(Tshirts.income)
print(Tshirts.pieces_sold)
```
