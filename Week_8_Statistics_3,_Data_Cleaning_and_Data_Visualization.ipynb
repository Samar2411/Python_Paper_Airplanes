#Statistical Tests

A quick list will be put here:

A. Chi-Squared Test
B. Fisher Exact Test
C. McNemar Test



#The Chi-Squared Test

It is also known as Chi Squared Test of Independence. 

Like all tests, it has conditions.

For the Chi-Square Test, less than 20% of the counts should have a value less than 5.

Observations are independent.


Lets say for example that we have two groups, one male and one female. We poll them to see what their preferred type of candy was, chocolate, licorice or pastries. Is there a difference between men and women in terms of their choice of sweets.


```python
data = [[120, 90, 40],
        [110, 95, 45]]
table = pd.DataFrame(data, ['male', 'female'], ['chocolate', 'licorice', 'pastries'])
table
```

The formula returns a ton of data but the p value, is the second output and the value is greater than .05 so there is no statistical significance. 

```python
import scipy.stats as stats
stats.chi2_contingency(data)
```

#Fisher Exact Test

Fischer Exact test is like the Chi Squared Test of Independence in that you compare discrete data but only when more than 20% of the counts are less than 5, you need to use Fisher Exact Test.

Lets say you had a clinical trial where you wanted to observe the effect of a drug on curing patients with diptheria. The patient samples are small, less than 5 in more than 20% of categories, so we need to use Fisher Exact test.

```python
data = [[6,2],
        [1,4]]
table = pd.DataFrame(data, ['sick', 'cured'], ['drug', 'sugar_pill'])
table
```

The second value is the p value so it is greater than .05 and therefore we conclude there is no significance between the two groups.

```python
from scipy.stats import fisher_exact
fisher_exact(data)
```

#McNemars Test

McNemars test is when you compare discrete data for a before and after group or paired group. 




```python
data = [[30,40],
        [12,18]]
table = pd.DataFrame(data, ['support', 'oppose'], ['no campaign ad', 'campaign ad'])
table
```

```python
from statsmodels.stats.contingency_tables import mcnemar
print(mcnemar(data))
```

#Subplot 

A lot of advanced data visualization customization options rely on the fig, ax = plt.subplots() syntax. The function returns a figure object and a tuple containing axes objects equal to nrows*ncols. 

What this does is place the figure into fig and the axes settings into ax. Once you do, there are many methods you can call on these two variables.





```python
x = np.arange(1,11)
y = np.random.randint(1, 100,10)
z = np.random.rand(10)
```

# Setting the Plot



```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax[0].plot(x,y)
ax[1].plot(x, z)
ax[0].set_title('Plot 1') 
ax[1].set_title('Plot 2')
```

# Plot inside a plot

The 4 data sets correspond to left, bottom, width, height of the plots. The values are from 0 to 1.

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([.1,.1,.8,.8])
ax2 = fig1.add_axes([.2,.5,.4,.3])
```

```python
ax1.plot(x,y)
ax2.plot(x, z)
fig1
```

# Changing figure size

```python
figure, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (15,5))
axes[0].plot(x,y)
axes[1].plot(x, z)
figure
```

```python
figure, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (3,2))
axes[0].plot(x,y)
axes[1].plot(x, z)
figure
```

# We can set a legend

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(x,x**2, label = 'square')
axes.plot(x, x**.5, label = 'square root')
axes.legend(loc = 10)
```

# Changing colors, linewidth, markers and styles



```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(x, y, color = 'red', linewidth = 5, linestyle = ':', marker = '1', markersize = 40, markeredgecolor = 'green')
```
