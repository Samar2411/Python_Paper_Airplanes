#Creating Normalized Data

np.random.normal creates randomly distributed data. The first argument is the mean, the second is how spread out you want it to be and the third argument is how many numbers you want.

# Hypothesis Testing

The field of statistics concerned with drawing conclusions or testing hypothesis about data which is an important skill in data science. Almost certainly, you will need to use a form of hypothesis testing statistical tool in your future. 

To determine which test we want to do, we need to answer a few questions.

1. How many groups are you comparing?
2. Is the data normally distributed?
3. What kind of data are you comparing? Is it discrete or continuous?
4. Are the samples random?
5. Are the samples independent?

Once you select your test, you now have a null hypothesis or alternate hypothesis. The null hypothesis is the conclusion you draw if there are no differences. It varies by test. The alternate hypothesis is usually a statement of change.

# One Sample T Test

To evaluate, the data needs to be 
  1. A normally distributed sample

    a. with unknown standard deviation and
    
    b. with a  number of samples n<30
  2. random samples 
  3. independent observations
  4. Continuous

The null hypothesis is that the means of the sample from the group is the same. The alternate hypothesis is that they are different in some way. 

Lets say we have a factory that produces cheese. On average the weight of the cheese blocks is 50grams. If we select randomly 20 peices of cheese and find that their average is 44.7. Are these data statistically different?

The first output is the calculated value and the second is the p value. If it is less than .05, we assert a difference.

In this case, we use the scipy package and the inputs are the dataset of samples. and the average we are comparing to. 

```python
import scipy.stats as stats
cheese = np.random.normal(40,50, 20)
print(cheese.mean())
print(len(cheese))
```

```python
stats.ttest_1samp( cheese , 50 )
```

# Two Sample T Test Unpaired

The two sample t test is nearly indentitical to the one sample t test but now we are comparing the means of two different samples. 

Lets say that we want to know if average monthly income of McDonalds employees at two chains. We sample 25 employees.

The package is found is scipy. For now, we will set the parameer equal_var = true. The arguments are the two datasets.

```python
McDonalds1 = np.random.normal(5000, 1, 25)
McDonalds2 = np.random.normal(8000, 1, 25)

print(statistics.variance(McDonalds1))
print(statistics.variance(McDonalds2))
```

```python
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(McDonalds1, bins = 10)
print(shapiro(McDonalds1))
plt.show()
```

```python
import scipy.stats as stats 
stats.ttest_ind(McDonalds1, McDonalds2, equal_var = True)
```

# Two Sample T test Paired

This is similar to the unpaired two sample t test but in this case, the samples are usually the same individuals or objects before and after time. 

For the criteria, we say that they are dependent on each other.

Lets say for example that you want to test cancer volume in 16 patients before and after adding chemotherapy drugs. We want to know if the change in volume is significant. To do this we need to use the scipy package and the only inputs are the before dataset and after dataset.

```python
before = np.random.normal(110, 3, 16)
after = np.random.normal(90, 5, 16)
stats.ttest_rel(before, after)
```
